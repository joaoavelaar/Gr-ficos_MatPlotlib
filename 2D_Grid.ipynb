```python
# Importando bibliotecas 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Criando variável X
X = np.linspace(0,5,11)
```

```python
# Criando funções Y(X) e Z(X)
Y= np.power(X-2,2.0)
Z= np.exp(X/2.0)
```

```python
# Criando grid 
plt.plot(X,Y,label= '$Y(X)=(X-2)^2$')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('y(x)')
plt.grid(visible=True , linestyle = '--', which = 'major', color = 'grey')
plt.legend()
plt.show()
```

```python
# Alterando cor do fundo
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.grid(linestyle = '--' , which = 'major' , color = 'black')
plt.legend()
plt.show()
```
