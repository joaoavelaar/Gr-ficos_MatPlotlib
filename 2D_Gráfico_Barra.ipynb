```python
#Importando bibliotecas e classes do projeto
import numpy as np # Trabalha com objetos  vetor e matriz multidimensional
import matplotlib.pyplot as plt # Gerar os gráficos de visualização
```

```python
# Criando os campos Descritivos (labels)
labels = ["A1","A2","A3","A4","A5"]
```

```python
# Criando os valores dos compos (Métrica do gráfico)
valores = [10,34,56,78,12]
```

```python
# Visualizando o gráfico com o objeto plt (Matplotlib).
plt.bar(labels,valores)
```

```python
# Visualizando o gráfico com o objeto plt (Matplotlib).
plt.barh (labels,valores)
```

```python
# Atribuindo novos aos eixos do gráfico (x, y).
plt.bar(labels,valores)
plt.xlabel('Codigo do Produto')
plt.ylabel('Valor do Produto')
```

```python
# Alterar as cores dos elementos gráficos.
plt.bar(labels,valores,color='gray')
plt.xlabel('Codigo do Produto')
plt.ylabel('Valor do Produto')
```

```python
# Colocando os valores existentes em cada tipo como rótulo de dados.
plt.bar(labels,valores,color='gray')
plt.xlabel('Codigo do Produto')
plt.ylabel('Valor do Produto')

for x,y in zip(labels,valores):
    plt.text(x, y+0.05, '%d' % y, ha='center' , va= 'bottom')
```

```python
# Criando um novo Vetor chamado valores2.
valores2 = [8,30,60,21,30]
```

```python
largura = 0.2
pos1 = np.arange(len(valores))
pos2 = [x + largura for x in pos1]

plt.bar(pos1,valores, color='gray', width=largura, label='Barra 1')
plt.bar(pos2, valores2, color='black', width=largura, label='Barra 2')
plt.legend()
```

```python
largura = 0.4
plt.bar(pos1,valores,color='black',width=largura,label = "Barra 1")
plt.bar(pos1,valores2,color='gray',width=largura,label = "Barra 2",bottom = valores)
```
