```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variável X 
X = np.linspace(-5,5,101)
```

```python
# Criando variável Y
Y= (X-2.0)*(X+2.0)
```

```python
# Graficando 
plt.plot(X,Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.fill_between(X,Y)
plt.show()
```
