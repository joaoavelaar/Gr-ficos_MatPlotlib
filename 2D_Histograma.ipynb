```python
# Importando biblioteca 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variável X com números aleatórios
X = np.random.rand(2000)
```

```python
# Alterando número de bins
plt.hist(X,bins=10)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bins=10')
plt.show()
```

```python
plt.hist(X,bins=20)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=20')
plt.show()
```

```python
plt.hist(X,bins=40)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=40')
plt.show()
```

```python
# Estilizando (Alterando cores)
plt.hist(X,bins=50,color='silver')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

```python
plt.hist(X,bins=50,color='silver', edgecolor='black')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

```python
# Estilizando (Multiplos histogramas na mesma imagem )
X1 = np.random.normal(0,1,size = 800)
X2 = np.random.normal(3,1,size = 800)
X3 = np.random.normal(5,1,size = 800)
```

```python
plt.hist(X1,bins=10,color='gray')
plt.hist(X2,bins=10,color='black')
plt.hist(X3,bins=10,color='orange')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Multiplos histogramas')
plt.show()
```

```python
# Estilizando (Distribuição cumulativa)
X = np.random.rand(4000)
```

```python
plt.hist(X,bins=20,color='silver', edgecolor='black',cumulative=True)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Histograma cumulativo')
plt.show()
```
