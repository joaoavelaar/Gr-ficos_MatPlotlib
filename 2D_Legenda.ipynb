```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variável X
X= np.linspace(0,5,11)
```

```python
# Criando variável Y(X) e Z(X)
Y= np.power(X-2,2.0)
Z= np.exp(X/2.0)
```

```python
# Colocando equações na legenda
plt.plot(X,Y,label='$Y(X)=(X-2)^2$', color= 'orange')
plt.plot(X,Z,label='$Z(X)=e^{(X/2)}$', color='black')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```

```python
# Alterando posição da legenda 
plt.plot(X,Y,label='$Y(X)=(X-2)^2$', color= 'orange')
plt.plot(X,Z,label='$Z(X)=e^{(X/2)}$',color='black')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend(loc='lower right')
plt.show()
```

```python
# Criando novas variáveis X, Y e Z
X= np.linspace(0,2.0*np.pi,361)

Y= np.cos(X/8)
Z= np.sin(X/5)

plt.plot(X, Y,label= '$\cos(X/2)$', color= 'orange')
plt.plot(X,Z, label='$\sin(X/4)$',color='black', linestyle='--')
plt.legend()
```
