```python
# Matplotlib - Gráfico de espalhamento - Parte 1/3
# Parte 1: Criando gráficos de espalhamento (scatter) com a biblioteca MatPlotLib
# João Antônio Avelar Duarte
```

```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variável X
X = np.linspace(0,5,11)
```

```python
#Criando funções Y(X) e Z(X)
Y = np.power(X-2,2.0)
Z = np.exp(X/2.0)
```

```python
#Gráfico de Y em função de X
plt.scatter(X,Y)
plt.show()
```

```python
#Inserindo nomes nos eixos
plt.scatter(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
# Inserindo título no gráfico
plt.scatter(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.show()
```

```python
# Dois gráficos na mesma imagem
plt.scatter(X,Y)
plt.scatter(X,Z)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
# Alterando cores
plt.scatter(X,Y,color='black')
plt.scatter(X,Z,color='orange')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
# Inserindo legenda
plt.scatter(X,Y,color='black',label='Y(X)')
plt.scatter(X,Z,color='orange',label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```

```python
# Parte 2/3: Alterando os simbolos no gráficos de scatter
```

```python
#Gráfico de Y em função de X
plt.scatter(X,Y)
plt.show()
```

```python
#Alterando símbolo
plt.scatter(X,Y,marker='.')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
# Dois gráficos com simbolos diferentes
plt.scatter(X,Y,marker='p')
plt.scatter(X,Z,marker='v')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
# Parte 3/3: Controlando tamanho do ponto
```

```python
# Gráfico de Y em função de X
plt.scatter(X,Y)
plt.show()
```

```python
# Alterando tamanho do ponto
plt.scatter(X,Y,s=15)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
plt.scatter(X,Y,s=15,color = 'black',label='Y(X)')
plt.scatter(X,Z,s=40,color = 'red',label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```

```python
plt.scatter(X,Y,s=15,color = 'black',label='Y(X)',alpha=0.8)
plt.scatter(X,Z,s=40,color = 'blue',label='Z(X)',alpha=0.4)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```
