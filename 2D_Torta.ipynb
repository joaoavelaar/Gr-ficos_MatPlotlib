```python
# Importando bibliotecas 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando as labels 
labels= ["A1","A2","A3","A4","A5"]
```

```python
# Criando lista de valores
valores = [10,45,32,56,23]
```

```python
# Criando gráficos de barras
plt.pie(valores, labels=labels)
plt.show()
```

```python
# Inserindo porcentagem
plt.pie(valores, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
# Explode 
explode = [0,0.1,0,0,0]
plt.pie(valores, labels=labels, autopct='%1.1f%%',explode=explode)
plt.show()
```
