```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando Variáveis X
X = np.linspace(9,6,30)
```

```python
# Criando Funções Y(X) e Z(X)
Y = np.power(X-5,3.0)
Z = np.exp(X/2.0)
```

```python
# Gráficos de Y em Função de X
plt.plot(X,Y)
plt.show()
```

```python
# Inserindo nomes nos eixos:
plt.plot(X,Y)
plt.xlabel ('Eixo X')
plt.ylabel ('Eixo Y')
plt.show()
```

```python
# Inserindo títulos no gráfico
plt.plot (X,Y)
plt.xlabel ('Eixo X')
plt.ylabel ('Eixo Y')
plt.title('Y(X)')
plt.show()
```

```python
# Dois gráficos na mesma imagem 
plt.plot(X,Y)
plt.plot(X,Z)
plt.xlabel ('Eixo X')
plt.ylabel ('Eixo Y')
plt.title ('Y(x), Z(X)')
plt.show()
```

```python
# Alterando cores
plt.plot(X,Y,color = 'black')
plt.plot(X,Z,color = 'green')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
# Inserindo legenda
plt.plot(X,Y,color = 'black',label= 'Y(X)')
plt.plot(X,Z,color = 'green',label= 'Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```
